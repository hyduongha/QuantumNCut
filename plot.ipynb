```python
import matplotlib.pyplot as plt
import numpy as np


# 4	1	0.237231512	0.369462132
# 4	2	6.76E-05	0.000249362
# 8	1	1.914871996	1.069105021
# 8	2	0.195226503	0.259679004
# 8	3	0.000332083	0.000695074
# 16	1	11.35062147	2.855070774
# 16	2	3.948745652	1.304608256
# 16	3	0.302854721	0.212198066
# 16	4	0.014104899	0.017399202



mean_data = [
	[0.237231512,6.76E-05],
	[1.914871996,0.195226503,0.000332083],
	[11.35062147,3.948745652,0.302854721,0.014104899]
]


var_data = [
	[0.369462132,0.000249362],
	[1.069105021,0.259679004,0.000695074],
	[2.855070774,1.304608256,0.212198066,0.017399202]
]


mean_data_qiskit = [
	0.013858575660905743,
	0.06411989329224471,
 	0.28488739376597305,
	1.2022788635410384,
	4.803528132427841
]

std_data_qiskit = [
	0.012689387666102612,
	0.04712762079828724,
	0.09836636280631506,
	0.27603784203796794,
	0.9030436174072689
]

fig, ax = plt.subplots(1,2, figsize=(8, 2))

ax[0].plot([1,2], mean_data[0], 'o-', label='2-qubit')
ax[0].fill_between([1,2], 
				np.array(mean_data[0]) - np.array(var_data[0]), 
				np.array(mean_data[0]) + np.array(var_data[0]), 
				alpha=0.2)


ax[0].plot([1,2,3], mean_data[1], 'o-', label='3-qubit')

ax[0].fill_between([1,2,3], 
				np.array(mean_data[1]) - np.array(var_data[1]), 
				np.array(mean_data[1]) + np.array(var_data[1]), 
				alpha=0.2)
ax[0].plot([1,2,3, 4], mean_data[2], 'o-', label='4-qubit')
ax[0].fill_between([1,2,3, 4], 
				np.array(mean_data[2]) - np.array(var_data[2]), 
				np.array(mean_data[2]) + np.array(var_data[2]), 
				alpha=0.2)
ax[0].set_xticks([1,2,3,4])

ax[0].set_xlabel('Number of Layers')
ax[0].set_ylabel('Loss')
ax[0].grid(True, which='both', linestyle='--', linewidth=1)

ax[0].legend()



ax[1].plot([2,3,4,5, 6], mean_data_qiskit, 'ro-', label='Qiskit')
ax[1].fill_between([2,3,4,5, 6],
				np.array(mean_data_qiskit) - np.array(std_data_qiskit), 
				np.array(mean_data_qiskit) + np.array(std_data_qiskit), 
				color = 'red', alpha=0.2)
ax[1].set_xticks([2,3,4,5, 6])

ax[1].set_xlabel('Number of Qubits')
ax[1].set_yticklabels([])
ax[1].grid(True, which='both', linestyle='--', linewidth=1)
ax[1].legend()


plt.tight_layout()
plt.savefig('./images/cost_vs_layers.svg', bbox_inches='tight')
```
