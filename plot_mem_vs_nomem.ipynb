```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
num_qubits = [3,4,5]
k = list(range(2, 20))
labels = ['(a)', '(b)', '(c)']
fig, axs = plt.subplots(3, 1, figsize=(5, 10))
for i, n in enumerate([3,4,5]):

    data = pd.read_csv(f'./data/mem_{n}qubit.csv', delimiter=',')
    ax1 = axs[i]
    ax2 = ax1.twinx()  # Create a twin y-axis sharing the same x-axis

    # Plot cost on the left y-axis
    ax1.plot(k, data['cost_with_mems'], label='Cost (proposed)', color='blue')
    ax1.plot(k, data['cost_without_mems'], label='Cost', color='blue', linestyle='--')
    
    ax1.tick_params(axis='y', labelcolor='blue')

    # Plot i on the right y-axis
    ax2.plot(k, data['i_with_mems'], label='Total. #Iter (proposed)', color='red')
    ax2.plot(k, data['i_without_mems'], label='Total. #Iter', linestyle = '--', color='red')
    ax2.tick_params(axis='y', labelcolor='red')
    ax1.set_ylabel('Cost', color='blue')
 
    ax2.set_ylabel('Total. #Iter', color='red')

    ax1.set_yticks([0.0007, 0.0009, 0.0011, 0.0013])
    #ax2.set_yticks(range(60, 200, 20))
    #ax1.set_yticklabels([0.0007, 0.0009, 0.0011, 0.0012, 0.0013])
    ax1.set_ylim(0.0007, 0.0014)
    #ax2.set_ylim(50, 200)
    # Set common x-axis label and title
    if i == 0 or i == 1:
        ax2.set_yticks([50, 60, 70, 80])
    else:
        ax2.set_yticks([100, 110, 120, 130, 140, 150])
    axs[i].set_xticks([2,4,6,8,10,12,14,16,18])
    if i == 2:
        axs[i].set_xlabel(rf'$k$')
        axs[i].set_xticklabels([2,4,6,8,10,12,14,16,18])
    else:
        axs[i].set_xticklabels([])
    ax1.set_title(f'{labels[i]} {n} qubits')

    # Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.tight_layout()
plt.savefig('./images/mem_cost_i.svg', bbox_inches='tight')
```
